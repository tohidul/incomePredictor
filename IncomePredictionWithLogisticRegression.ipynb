```python
b = np.ones((df_t.shape[0],1))
```

```python
data_train_array = np.column_stack([b, data_train_array])
```

```python
y = data_train_array[:,4]
```

```python
data_train_array = np.delete(data_train_array, [4], axis=1)
```

```python
data_train_array.shape[1]
```

```python
theta = np.ones(data_train_array.shape[1])
```

```python
thetaTemp = theta
```

```python
b_v = np.ones((df_v.shape[0],1))
```

```python
data_validation_array = np.column_stack([b_v, data_validation_array])
```

```python
y_v = data_validation_array[:,4]
```

```python
data_validation_array = np.delete(data_validation_array, [4], axis=1)
```

```python
iteration = 30
alpha = 0.01
```

```python
max_val_acc = 0
for abc in range(iteration):
    for j in range(data_train_array.shape[1]):
        temp=0
        for i in range(data_train_array.shape[0]):
            thetax = np.dot(theta, data_train_array[i])
            hx = 1/(1+exp(-thetax))
            temp = temp + (alpha*((hx-y[i])*data_train_array[i][j]))
            #print(temp)
        
        thetaTemp[j] = theta[j] - temp
        temp = 0
    
    correct = 0;
    incorrect = 0;
    correct_v = 0;
    incorrect_v = 0;
    
    for k in range(data_train_array.shape[0]):
        thetax = np.dot(theta, data_train_array[k])
        hx = 1/(1+exp(-thetax))
        if(hx<0.5):
            x = 0
        else:
            x = 1
        if(y[k]==x):
            correct = correct+1
        else:
            incorrect = incorrect +1
    for l in range(data_validation_array.shape[0]):
        thetax = np.dot(theta, data_validation_array[l])
        hx = 1/(1+exp(-thetax))
        if(hx<0.5):
            x = 0
        else:
            x = 1
        if(y_v[l]==x):
            correct_v = correct_v+1
        else:
            incorrect_v = incorrect_v +1
    
    val_acc = (correct_v/(correct_v+incorrect_v))*100
    if(val_acc>max_val_acc):
        max_val_acc = val_acc
        best_theta = theta
    print("After Iteration", abc, "\n")
    theta = thetaTemp
    print(theta, "\n")
    print("Train Accuracy: ",((correct/(correct+incorrect))*100), "\n")
    print("Validation Accuracy: ",val_acc, "\n")
```

```python
print("Max Validation Accuracy: ",max_val_acc, "\n")
print("Best Theta: ",best_theta, "\n")
```
